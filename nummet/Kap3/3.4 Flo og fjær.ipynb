```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,24,200)
y = 3.2 * np.sin(np.pi/6*(x-3))
```

```python
plt.rcParams.update({'font.size':12}) #all teksten i plottet vil ha skrifttypestørrelse på 15. 
plt.xlabel('t [timer]') #navn på x-aksen
plt.ylabel('T [meter]') #navn på y-aksen
plt.plot(x,y,linewidth=3, label='T(t)')
plt.hlines(-1,0,-24,'k', label='y=-1')
plt.plot(np.array([6,18]), np.array([3.2,3.2]),'ro',linewidth=2,label='Flo') #legger inn de røde sirkleneplt.plot(np.array([2.39,9.61,14.39,21.61]),-np.ones([4,1]), 'md', linewidth=2, label='T(t)=-1')
plt.plot(np.array([2.39,9.61,14.39,21.61]),-np.ones([4,1]), 'md', linewidth=2, label='T(t)=-1') #legger inn de "diamantene" #'np.ones([4,1])' lager en vektor med 4 elementer som alle er 1
plt.axis([0,27,-4,4])
plt.grid(visible=True)
plt.legend(loc='upper right')
plt.show()
```
