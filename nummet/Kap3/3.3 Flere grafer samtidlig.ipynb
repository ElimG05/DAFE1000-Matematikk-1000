```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,1000) #a)
```

```python
f=np.cos(np.pi*x/20)**2 * np.sin(3*x)
```

```python
plt.plot(x,f)
plt.show()
```

```python
g=np.cos(np.pi*x/20)**2 #b)
```

```python
plt.plot(x,f)
plt.plot(x,g)
plt.axis([4,8,-.7,.7])
plt.show()
```

```python
x=np.linspace(0,2,1000) #d)
```

```python
a=np.cos(2*np.pi*x)
```

```python
b=np.cos(np.pi*x)
```

```python
c=2*np.cos(2*np.pi*x) 
```

```python
d=np.cos(np.pi*(x**2))
```

```python
plt.plot(x,a,'k',label='a(x)')
plt.plot(x,b,'r--',label='b(x)')
plt.plot(x,c,'b:',label='c(x)')
plt.plot(x,d,'g-.',label='d(x)')
plt.legend(loc='upper right')
plt.show()
```
