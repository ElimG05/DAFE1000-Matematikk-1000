```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
help(np.linspace)
```

```python
x=np.linspace(0, 7*np.pi, num=200, endpoint=True, retstep=False, dtype=None, axis=0) #a)
```

```python
y=np.sin(x) #b)
```

```python
plt.plot(x,y) #c)
```

```python
plt.plot(x,y,'r') #d)
```

```python
plt.plot(x,y,linewidth=3)
```

```python
plt.plot(x,y,'g--')
```
