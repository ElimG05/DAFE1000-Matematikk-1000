```python
#importere de pyhon bibliotekene jeg trenger:
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definerer funksjonen f(x), den vi fikk i oppgaven:
def f(x):
    return np.exp(-x/4) * np.arctan(x)
```

```python
#Definerer den deriverte av f(x) som en ny funksjon g(x):
def g(x):
    return np.arctan(x) - (4/(x**2+1))
```

```python
#Definerer den deriverte av g(x), dette er det samme som den dobbel deriverte av f(x):
def dg(x):
    return 1/(x**2+1) + 8*x/(x**2+1)**2
```

```python
x = 1 #Velger 1 som min start verdi
pres = 1e-8 #Setter presisjonen til 
iterasjoner = 0 #Setter antall iterasjoner lik 0
x_gml = 200 #Velger et tilfeldig tall inn for "gamle" x-verdi
```

```python
while abs(x - x_gml) > pres: #Løkken kjører så lenge denne betingelsen er oppnådd
    x_gml = x #Kopierer den gamle x
    
    #Iterasjonsformel
    x = x - (g(x))/(dg(x))
    iterasjoner += 1 #Antall iterasjoner øker etterhvert som løkken kjører
    print(f'x = {x:.5f}') #Printer ut x-verdien etterhvert som den blir regnet på
print(f'Iterasjoner = {iterasjoner}') #Printer ut antall iterasjoner vi har gjort etter at løkken avsluttes
```

```python
#Jeg får x = 1.69071 flere ganger, med fire rette desimaler, dette blir da den beste tilnærmingen til toppunktet i funksjonen f(x): 
maksimalpunkt = x #Setter x-verdien jeg fant som 'maksimalpunkt'
y_verdi = f(x) #Finner y-verdien til toppunktet 
```

```python
x = np.linspace(1,2,500) 
plt.plot(x,f(x), 'b', label="f(x)") #Plotter f(x)
plt.scatter(maksimalpunkt,y_verdi, color='red', label="maksimalpunkt") #Merker toppunktet til funksjonen:
plt.legend()
plt.grid() #Legger til rutenett
```
