a) i) Vi trenger Python eller et annet lignende verktøy for å kunne løyse likningen: sqrt.x = cos(x), de gjør arbeidet for oss, noe som tar mye kortere tid enn om vi skulle gjort det for hånd

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2,100) 
```

```python
a=np.sqrt(x)
b=np.cos(x)
```

```python
plt.plot(x,a,'r',label='sqrt(x)')
plt.plot(x,b,'b',label='cos(x)')
plt.legend(loc='lower left')
plt.show()
```
