iii) I likningen vår sqrt(x)= cos(x) har vi kun én variabel, x. Vi kan dermed flytte alt på en side og få noe slikt: f(x) = sqrt.x - cos(x). f(x) = 0 blir i vårt tilfelle ca. lik 0.64.

```python
#b) #Gresnser
a = 0
b = 2
#Her er a<b og f(a) og f(b) har ulike fortegn
```

```python
def f(x):
    return(np.sqrt(x) - np.cos(x)) #funksjonen vår
```

```python
fa = f(a)
fb = f(b)
```

```python
for i in range(0,11): #For-lækka kjører 10 ganger
    c = (a+b)/2 #f(c) er midtpunktet mellom a og b
    fc = f(c)
    if fa * fc < 0: #Hvis f(a) * f(c) blir et negativt tall, betyr det at de har motsatt fortegn og blir c vår nye b (høyre grense)
        b = c
    else:
        a = c #Hvis ikke så har de samme fortegn, så vi bare lar c bli vårt nye a (venstre grense)

    print(f'c = {c:.3f}')
```
